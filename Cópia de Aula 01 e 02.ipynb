A nota final de um estudante é calculada a partir da média de três notas atribuídas entre o intervalo de 0 até 10, respectivamente, a um trabalho de laboratório, a uma avaliação semestral e a um exame final. De acordo com o resultado, mostre na tela se o aluno está reprovado (média <= 4), de recuperação (4 <= media < 7) ou se foi aprovado (media >= 7). Ao final informa se o aluno está aprovado, em recuperação ou reprovado.

```python

tl = float(input("Digite sua nota no trabalho de laboratório: "))
aval = float(input("Digite sua nota na avaliação semestral: "))
examf = float(input("Digite sua nota no exame final:"))
media = float(tl + aval + examf)/3
if media <= 4:
  print("reprovado")
elif media <=4 and media <7:
  print("recuperação")
elif media >= 7:
  print("aprovado")
```

FGTS
Desconto de INSS
Desconto de imposto de renda
Salário líquido.
	
Informações:
	
	
	

*   O valor do FGTS é 8% do salário bruto
*   O valor do INSS é de 12% do salário bruto
* O valor do imposto de renda segue a seguinte regra:
* Se o salário bruto for menor que 1500, desconta 7,5%
* Se for entre 1500 e 2000, desconta 10%
* Se for acima de 2000, desconta 15%
* O salário líquido é igual ao salário bruto, descontados o INSS e IR




```python
imposto_de_renda = 0
salario_bruto = float(input("Digite seu Salario bruto"))
print("a. Salário bruto: ", salario_bruto)
FGTS = float((0.08 * salario_bruto))
print("b. FGTS: " , FGTS )
INSS = float((0.12 * salario_bruto))
print("c. Desconto do INSS " , INSS)

if salario_bruto < 1500:
  imposto_de_renda = 0.075 * salario_bruto
  print("d. Desconto do imposto_de renda " , imposto_de_renda )
elif salario_bruto >=1500 and salario_bruto <=2000:
  imposto_de_renda = 0.1 * salario_bruto
  print("d. Desconto do imposto de renda ", imposto_de_renda )
elif salario_bruto >2000:
  imposto_de_renda = 0.15 * salario_bruto
  print("d. Desconto do imposto de renda ", imposto_de_renda )

salario_liquido = float((salario_bruto - INSS))
print("e. Salário liquido: ", salario_liquido)
	
```

Você é programador de um escritório de contabilidade e tem a missão de desenvolver uma aplicação que elabora o contracheque de cada funcionário. Elabore um algoritmo que recebe o salário bruto de um usuário e em seguida retorne na tela as seguintes informações:

* O salário bruto é: ___
* O recolhimento do FGTS será: ___
* O recolhimento de INSS será: ___
* O salário líquido este mês será: ___

OBS. O valor do recolhimento de FGTS é de 8% do salário bruto, o recolhimento de INSS é de 12% do salário bruto e o salário líquido é igual ao salário bruto - o INSS.


```python
salario_bruto = float(input("Digite seu salário bruto: "))
print ("O salário bruto é:", salario_bruto) 

FGTS = float((0.08 * salario_bruto))
print ("o recolhimento do FGTS será: ", FGTS)

INSS = float(0.12 * salario_bruto)
print("O recolhimento do INSS será ", INSS)

salario_liquido = float((salario_bruto-INSS))
print("O salário líquido este mês será", salario_liquido)
```
